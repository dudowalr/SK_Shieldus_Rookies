```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 열을 추가 : 시군구
# 구 별로 스타벅스 매장이 몇개가 있는지 
data_star4.loc[:, '시군구'] = data_star4['소재지전체주소'].str.split().str[1]

data_star4
```

```python
data_5 = data_4.groupby('시군구').size().reset_index(name='커피숍')
data_5
```

```python
data_star5 = data_star4.groupby('시군구').size().reset_index(name='스타벅스')
data_star5
```

```python
# 시군구로 merge하기
df = pd.merge(data_5, data_star5, on='시군구')
print(df)

#비율로 보기
df['스타벅스비율'] = df['스타벅스'] / df['커피숍'] * 100
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

fig, ax1 = plt.subplots(figsize=(12,6))

# 커피숍수 (막대)
ax1.bar(df['시군구'], df['커피숍'], color='gray')
ax1.set_ylabel('커피숍 수')

# 스타벅스 점유율 (선)
ax2 = ax1.twinx()
ax2.plot(df['시군구'], df['스타벅스비율'], marker='o', color='black')
ax2.set_ylabel('스타벅스 점유율 (%)')

plt.setp(ax1.get_xticklabels(), rotation=45, fontsize=8)

plt.title('시군구별 커피숍 수와 스타벅스 점유율')

plt.show()
```

커피숍이 많은 지역일수록 스타벅스 점유율도 상대적으로 높게 보이며, 커피숍 수가 적은 지역에서는 스타벅스 점유율이 낮은 경향이 있다. 

스타벅스 비중이 높은곳은 유동인구가 많은 상업지역이라 볼 수 있고 적은곳은 상업활동 규모가 작은 지역일 가능성 유추 가능
