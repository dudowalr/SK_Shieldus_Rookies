1.숫자를 입력 받아 짝수, 홀수를 판별하는 프로그램

```python
input_num = int(input('숫자를 입력하세요: '))

if input_num % 2 == 0:
    print(f'{input_num} , 짝수입니다')
else:
    print(f'{input_num} , 홀수입니다')
```

2. 숫자를 이벽 받아 양수, 음수, 0을 판별하는 프로그램

```python
input_num = int(input('숫자를 입력하세요: '))

if input_num > 0:
    print(f'{input_num} , 양수입니다')

if input_num < 0:
    print(f'{input_num} , 음수입니다')

if input_num == 0:
    print(f'{input_num} , 0입니다')
```

3. 독감예방 접종이 가능한지 여부를 확인하는 프로그램

```python
input_num = int(input('나이를 입력하세요: '))

if (input_num < 15) or (input_num >= 65) :
    print(f'{input_num}세 , 무료 입니다.')

else :
    print(f'{input_num}세 , 유료 입니다.')
```

4. 특정 연도 건강검진 대상 여부 판별 및 검진 종류 확인
