```python
import csv # csv 생성, 저장 모듈
import re # 정규표현식 모듈

class LogAnalyze:
    def __init__(self): # 생성자(객체가 생성될때 실행)
        self.all_ip_list = [] # ip주소 저장 리스트 

    def analyze(self): # log 읽고 ip 추출
        file_path = input('로그 파일 경로 입력: ')

        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                error_log = False 
                # IPv4 형식 검증식
                pattern = r'^(?:(?:25[0-5]|2[0-4][0-9]|1?[0-9]{1,2})\.){3}(?:25[0-5]|2[0-4][0-9]|1?[0-9]{1,2})'
                for line in file:
                    match = re.match(pattern, line) #로그 비교
                    if match:
                        ip = match.group() 
                        self.all_ip_list.append(ip) # 리스트에 저장
                    else: # 잘못된 로그 한번만 출력
                        if not error_log:
                            print('잘못된 형식')
                            error_log = True

            self.result_display() # 출력, 저장

        except FileNotFoundError:
            print('파일을 찾을 수 없습니다.')
        except PermissionError:
            print('읽기 권한이 없습니다')
        except Exception as e: # 나머지
            print(f'오류 발생: {e}')
    
    def save_csv(self, count):
        try:
            # csv 저장시 newline=''추가 (무결성 관리)
            with open('ip_frequency.csv', 'w', newline='',  encoding='utf-8') as file:
                write = csv.writer(file)
                write.writerow(['IP Address', 'Frequency'])

                for ip, frequency in count.items():
                    write.writerow([ip, frequency])
                print('파일 저장 완료')

        except Exception as e:
            print(f'저장중 오류 발생: {e}')

    #출력, 저장 
    def result_display(self):
        if not self.all_ip_list:
            print('ip주소가 없습니다.')
            return
        
        # 빈도계산
        ip_count = {}
        for i in self.all_ip_list:
            if i in ip_count:
                ip_count[i] += 1
            else:
                ip_count[i] = 1

        #빈도 상위 3개 ip정렬
        precedence_top_3 = sorted(ip_count.items(), key=lambda x: x[1], reverse=True)[:3]
        # 출력
        for i, (ip, frequency) in enumerate(precedence_top_3, 1):
            print(f'빈도수 {i}위: {ip} : {frequency}회')
        
        #출력 끝나면 csv 저장
        self.save_csv(ip_count)

    def main():    
        analyzer = LogAnalyze()
        analyzer.analyze()

    # 실행
    main()
```
