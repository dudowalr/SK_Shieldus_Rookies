```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
# 데이터 불러오기
df = pd.read_csv('./dataset/diabetes.csv')
```

데이터셋

```python

# 데이터 확인
print(f'데이터 내용')
print(f"{'-'*60}")
print(f'{df.info()}')
print(df.describe())

```

결측치 처리
 Glucose, BloodPressure, SkinThickness, Insulin, BMI 열에서 0을 결측치로 간주하고 평균으로 대체

```python
# 결측치로 간주하고 평균으로 대체할 열 목록
data_cols_1 = [
    'Glucose', 
    'BloodPressure',
    'SkinThickness',
    'Insulin',
    'BMI'
    ]

# 열에서 0을 결측치로 간주하고 평균으로 대체  
print(f'결측치 처리전 데이터 출력(0 포함)')
print(f"{'-'*60}")
print(f'{df[data_cols_1]}')
print(f"{'-'*60}\n")

# 결측치 확인
df[data_cols_1] = df[data_cols_1].replace(0, np.nan) # 0 값을 우선 NaN으로 변경
df[data_cols_1] = df[data_cols_1].fillna(df[data_cols_1].mean()) # NaN을 평균으로 대체

print(f'결측치 처리 후')
print(f"{'-'*60}")
print(df[data_cols_1]) # 0자리에 평균값이 채워진것을 확인 가능

```

이상치 처리
SkinThickness, Insulin 상위 1%를 이상치로 간주하고 평균으로 대체

```python
data_cols_2 = [
    'SkinThickness',
    'Insulin'
    ]
    
for i in data_cols_2:
    boundary_data = df[i].quantile(0.99) # 99% 경계값 계산
    current_mean = df[i].mean()
    # 상위 1% 이상치를 NaN으로 변경
    df.loc[df[i] > boundary_data, i] = np.nan
    # NAN 값을 해당 열의 평균으로 채움
    df[i] = df[i].fillna(current_mean)

# SkinThickness, Insulin 상위 1%를 평균으로 대체 후 데이터 확인 
print(df[data_cols_2])
```

정규화 Age 열을 0~1 범위로 정규화

```python
scaler = MinMaxScaler()
data_cols_3 = ['Age']
df[data_cols_3] = scaler.fit_transform(df[data_cols_3])

# 정규화 유무 확인
print(f'정규화 확인')
print(f"{'-'*15}")
print(f'{df[data_cols_3]}')

```

EDA

```python
# EDA
# 각 열의 결측치 개수 출력
missing_cols = df.isnull().sum()
print(f'각 열의 결측치 개수 출력')
print(f"{'-'*40}")
print(f'{missing_cols}')
print(f"{'-'*40}\n")

#Outcome 별 Glucose 평균 출력
outcome_data = df.groupby('Outcome')['Glucose'].mean()
print(f'Outcome 별 Glucose 평균 출력')
print(f"{'-'*40}")
print(f'{outcome_data}\n')
print(f"{'-'*40}\n")

#전처리 후 데이터프레임 상위 5개행 출력
top5_data = df.head(5)
print(f"{'-'*40}")
print(f'상위 5개행')
print(f'{top5_data}')
print(f"{'-'*40}")
```
