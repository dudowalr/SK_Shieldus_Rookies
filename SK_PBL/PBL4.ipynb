```python
#PBR4 김재영

class StudentScores: #(클래스)
    # 생성자
    def __init__(self, txt_file_name): 
        self.scores = {} # 학생 이름과 점수를 저장하는 딕셔너리(속성)  
        try:
            with open(txt_file_name, 'r', encoding = 'utf-8') as file:# with => close() 메서드를 생략
                for i in file:
                    student_value = i.split(',') #','기준 리스트로 만들기
                    student_name = student_value[0] # list 0 = 이름
                    student_score = student_value[1] # list 1 = 점수
                    self.scores[student_name] = int(student_score) # 점수를 int형으로 바꿔서 dict로 넣기
        except ValueError: 
            print('데이터가 올바르지 않습니다.')
        except FileNotFoundError: 
            print('파일을 찾을 수 없습니다.')
        
    # 평균점수 계산(메서드)
    def get_average(self):
        total = sum(self.scores.values()) # 점수의 합
        average = total / len(self.scores) # 점수의 합 / 딕셔너리 길이
        return average 

    # 평균 이상 학생 이름 리스트 반환(메서드)
    def above_average(self):
        result = []
        average = self.get_average()
        for name, score in self.scores.items():
            if score >= average: #평균 이상일떄
                result.append(name) #리스트(result) 마지막 요소로 추가
        return result

    # 평균미만 학생 파일 생성, 저장(메서드)
    def save_under_average(self, file_name):
        average = self.get_average()
        try:
            with open(file_name, 'w', encoding = 'utf-8') as file:
                for name, score in self.scores.items(): # items() -> 딕셔너리에 있는키와값들의 쌍 
                    if average >= score: #평균이하일때
                        file.write(f'{name},{score}\n') #줄바꿈 저장

        except Exception as e:
            print(f'파일 저장중 에러발생: {e}')

    # 평균점수와 평균이상 학생 리스트 출력 (메서드)
    def print_result(self):
        average = self.get_average()
        above_students = self.above_average()
        print(f"{'--' * 30}")
        print(f'평균 점수 : {round(average, 1)}') # 소수점 첫째까지 나오게 반올림
        print(f'평균 이상을 받은 학생들 : {above_students}')
        print(f"{'--' * 30}")


# 프로그램 실행
student = StudentScores("scores_korean.txt") #scores_korean.txt 읽기
student.save_under_average("below_average_korean.txt") # 평균미만 학생 파일로 저장
student.print_result()  # 결과 출력
 
```
