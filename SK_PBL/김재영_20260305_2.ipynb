1. 1부터 100까지의 합을 구하는 프로그램 작성

```python
total = 0 # 누적값을 담을 변수 total초기화
for i in range(1, 101): #1 ~ 101 - 1까지
    total += i # total = 0 + 1, total = 1 + 2, total = 3 + 3 .... i = 100까지 반복  
print(total) # 출력
```

2. for문을 이용하여 1부터 10까지를 곱한 결과를 출력하는 프로그램 작성

```python
total = 1 # 곱 누적값을 담을 변수 total초기화
for i in range(1, 11): #1 ~ 11 - 1까지
    total *= i # total = 1 * 1, total = 1 * 2, total = 2 * 3, total = 6 * 4.... i = 10까지 반복  
print(total) # 출력
```

4. while문을 무한 루프로 구성하여 작성하는 예제

```python
total = 0
i = 1

while 1:
    if (i % 2 == 0) or (i % 3 == 0):
        total += i

    if total >= 1000:
        break

    i += 1

print(f'1000이 넘어가는 합: {i}')
print(f'1000이 넘어선 값: {total}')
```
