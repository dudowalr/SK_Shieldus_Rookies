상품 기본 정보 만들기 (단가, 수량 -> int, 이름 -> str)

```python
# 우유
milk_name = '우유'
milk_price = 2300
milk_qty = 2

# 빵
bread_name = '빵'
bread_price = 1800
bread_qty = 1

# 계란
egg_name = '계란'
egg_price = 5200
egg_qty = 1
```

장바구니 구성하기 (컬렉션 활용)

```python
# 각 상품 정보를 dict 형태로 생성(예:{"name": "우유", "price": "2300", "qty":2})
milk = {'name' : milk_name, 'price' : milk_price, 'qty' : milk_qty}
bread = {'name' : bread_name, 'price': bread_price, 'qty' : bread_qty}
egg = {'name': egg_name, 'price' : egg_price, 'qty' : egg_qty}

# 3개 딕셔너리를 list로 묶어 cart 변수에 저장
cart = [milk, bread, egg]
```

장바구니 금액 계산하기 (연산자 활용)

```python
# 반복문 없이 계산
# 각 상품의 총액(단가 * 수량)
milk_total = milk_price * milk_qty
bread_total = bread_price * bread_qty
egg_total = egg_price * egg_qty

# 총 상품 금액 (세 상품의 총액 합계)
subtotal = milk_total + bread_total + egg_total

# 배송비 (조건 판단없이 3000원으로 고정)
delivery_fee = 3000

# 최종 결제 금액(총 상품 금액 + 배송비)
final_total = subtotal + delivery_fee

# 상품 1개당 평균 가격 (총 상품 금액 / 총 수량)
# 총 수량(total_qty = 우유 + 빵 + 계란)
total_qty = milk_qty + bread_qty + egg_qty
avg_price = subtotal / total_qty

# 할인 대상 여부 (최종 결제 금액이 10000원 이상이면 True, 아니면 False가 저장되도록 비교 연산 사용)
has_discount = final_total >= 10000
```

요약 문자열 만들기 (문자열 포멧팅)

```python
# summary 변수에 저장 후 출력
summary = f"\"장바구니: {milk_name} {milk_qty}개, {bread_name} {bread_qty}개, {egg_name} {egg_qty}개 | 상품합계 {subtotal}원 | 배송비 {delivery_fee}원 | 최종 {final_total}원 | 할인대상 {has_discount}\""

#출력
print(summary)
```
