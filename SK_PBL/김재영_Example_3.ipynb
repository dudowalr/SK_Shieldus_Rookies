```python
import pandas as pd
```

```python
df = pd.read_csv("Diamonds.csv")
df
```

```python
#IQP
Q1 = df['carat'].quantile(0.25)
Q3 = df['carat'].quantile(0.75)
IQR = Q3-Q1
print(Q1, Q3, IQR)
low_ = Q1 - (1.5 * IQR) # 하한
up_ = Q3 + (1.5 * IQR) # 상한
```

```python
# 이상치 값 찾기 
df_1 = df[ (df['carat']<low_) | (df['carat']>up_) ]
df_1
```

```python
# 이상치를 제거한 데이터
delet_data = df[(df['carat']>=low_) & (df['carat']<=up_)]
delet_data
```

```python
# 상한, 하한값 출력
print(f'상한값: {up_}')
print(f'하한값: {low_}')
# 이상치를 제거한 데이터 개수 출력
len(delet_data)
```
