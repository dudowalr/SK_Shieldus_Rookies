```python
# 1. 아래 메뉴와 가격표를 딕셔너리(dict)로 준비
menu = {'아메리카노' : 3000, '라떼' : 4000, '녹차' : 4500, '케이크' : 5500}

# 2. 주문 받은 메뉴를 차례대로 저장할 빈 리스트 생성(예: orders = [])
orders = []

# 3. 주문 입력 받기
while True: # while 반복문으로 주문을 계속 입력 받기
    order = input('메뉴를 입력하세요(끝 입력 시 종료): ') # "메뉴를 입력하세요(끝 입력시 종료):" 출력 후 입력
    if order == '끝': # 입값이 끝이면 반복 종료
        break
    if order in menu:            #입력한 메뉴가 가격표(dict)에 있으면 order(list)에 추가
        orders.append(order)
    else:                        # 가격표에 없으면 "없는 메뉴입니다." 출력 후 다시 입력
        print('없는 메뉴입니다.')

# 4 합계 계산하기
# 주문 종료 후 orders에 담긴 메뉴를 이용해 총 합계(total) 계산
# 주문 개수도 함께 구함(len() 사용 가능)
orders_count = len(orders)

# 총 금액 계산
total = 0                # 총 금액
for i in menu:
    count = 0
    for j in orders:
        if j == i:
            count += 1
    total += count * menu[i]
    print(i, count)

# 5. 할인 규칙 적용 if분기문 (20000원 이상이면 10% 할인)
# 변수 초기화(초기화 하지 않으면  20000미만에서 오류발생 )
discount_price = 0
last_payment = total

if total >= 20000:
        discount_price = int(total * 0.1)     # float -> int형 변환(출력 int로 하기 위해)
        last_payment = total - discount_price # 최종결제 가격

# 6. 주문 요약 출력
print(f'주문 목록: {orders} 총 주문 수: {orders_count} 메뉴 종류: {set(orders)} 합계: {total}원 할인: {discount_price}원 최종 결제: {last_payment}원')
```
