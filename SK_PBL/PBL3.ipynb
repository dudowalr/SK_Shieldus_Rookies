```python
# PBL3, 김재영
# 함수 설계 (권장) 아래와 같이 역할을 쪼개서 함수를 만들어보세요.(함수 이름은 변경해도 무방합니다.)

# 학생 수를 입력받고 학생수에 맞춰 학생 이름과 학생 점수 딕셔너리 생성하는 함수
def get_students():
    data = {}
    while 1:
        try:
            students_num = int(input('학생 수를 입력해주세요.')) # 학생수 입력 -> 정수 변환
            if students_num <= 0:                          # 0명 이하시 다시 입력 반복
                print('학생 수는 1명 이상 입력해야 합니다.')
                continue
            break
        except ValueError:                  # 숫자가 아닐경우 예외처리
            print('학생 수를 입력해주세요.')

    for i in range(students_num):
        students_name = input('학생 이름를 입력해주세요.') # 입력된 이름 문자열로 저장

        while 1:
            try:
                students_score = int(input('0~100까지 학생 점수를 입력해주세요.')) # 0~100까지 입력된 점수 정수로 저장
                if students_score < 0 or students_score > 100 : # 0보다 작거나 100보다 크면 다시입력
                    print('0~100 범위를 벗어났습니다 다시 입력해주세요.')
                    continue
                break

            except ValueError: # 숫자가 아닐경우 예외처리
                print('0~100까지 숫자를 입력해주세요.')

        data[students_name] = students_score # 입력된 이름, 점수 dict로 저장, 중복이름 => 기존 점수자리에 덮어쓰기

    return data # dict 반환

# 평균 계산
def calc_average(scores):
    total = sum(scores)

    average_total = round(float(total / len(scores)),2) # 평균값 계산 float로 저장

    return average_total

# 최고점수 및 최저점수
def find_top_bottom(student_dict):
    max_score = max(student_dict.items(), key=lambda x: x[1]) #문자열 비교로 결정
    min_score = min(student_dict.items(), key=lambda x: x[1])    
    
    return max_score, min_score # 최고, 최저 반환

# 수에 따른 등급 문자열 반환
def get_grade(student_dict):
    grade_dict = {} 
    for name, score in student_dict.items():
        if(score >= 90): # 90 이상 A
            grade = 'A'
        elif(score >= 80): # 80 이상 89 이하 B
            grade = 'B'
        elif(score >= 70): # 70 이상 79 이하 C 
            grade = 'C'
        elif(score >= 60): # 60 이상 69 이하 D
            grade = 'D'
        else:
            grade = 'F' #59 이하 F

        grade_dict[name] = grade # 등급(A~F) 딕셔너리에 저장

    return grade_dict

# 요약 리포트 최종 출력 
def print_report(student_dict):

    average_score = calc_average(list(student_dict.values())) # 평균값
    top_score, bottom_score = find_top_bottom(student_dict)# 최고, 최저값
    grades = get_grade(student_dict) # 등급

    print(f'=== 성적 리포트 ===')
    print(f'평균: {average_score:.2f} 최고: {top_score[0]}({top_score[1]})')# 소수점 2자리까지 출력
    print(f'최저: {bottom_score[0]}({bottom_score[1]})') #최저값 출력
    print('목록(이름순)')
    for name, score in sorted(student_dict.items()):# 이름 오름차순으로 정렬 및 출력
        print(f'□ {name} : {score}점 / {grades[name]}')
        
students  = get_students()
print_report(students)
        
```
