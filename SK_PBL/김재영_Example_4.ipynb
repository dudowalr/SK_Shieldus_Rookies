```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("adult.csv")
```

```python
encoder_1 = LabelEncoder()
col_encode = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
for i in col_encode:
    df[i] = encoder_1.fit_transform(df[i])
print(df.head())
```
