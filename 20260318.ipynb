```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X, y = make_blobs(
    n_samples=300,
    centers=4,
    random_state=42
)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
kmeans = KMeans(
    n_clusters=3,
    init='k-means++',
    random_state=42
)
```

```python
# 학습, 예측 -> fit_predict
cluster_labels = kmeans.fit_predict(X)
```
