(인터프리터 언어 특)변수 : 객체를 가리키는 것
특정 값을 기억하고 있다가 필요시점에 특정 값을 사용해야함.
파이썬에서는 num(변수) = 10(값, data, 리터럴)
10 데이터를 메모리 공간에 저장, 10이 저장된 메모리 공간 주소를 num이라고 하는 메모리 공간에 저장
변수는 주소값을 저장하는것이다.
변수는 메모리의 주소 값을 가짐. 
객체(저장된 데이터) = 자료형
id() 함수로 메모리 주소 확인
변수 선언 방법  -> 변수 이름 = 변수에 저장할 값(데이터)
문자와 숫자, _(underscore)사용가능, 대소문자 구분, 의미 있는 단어로 사용, 숫자로 시작하면 사용 불가, 공백이 포함될 수 없음. 
예약어를 변수명으로 사용할 수 없음(for, if True등)
파이썬은 예약어가 많으니 구분 잘 해야함. EX) list, for, tuple등

```python
# 변수 : 코드 내부에서 필요한 시점에 재사용하기 위해 데이터를 저장해놓는 방법
# 데이터는 어디에 저장? => RAM FLASH MEMORIE(전원이 종료되면 메모리의 데이터도 모두 소멸)
# 변수 => 식별자(구별하는것) : 값을 가르키는 역할, 특정 기능을 가르키는 역할. 숫자, 문자열, 논리값(참 True, 거짓 False), 객체(리스트, 튜플, 딕셔너리, 집합)
# # : 주석 : 프로그래밍 코드 안에서 메모 작성

num = 100
#변수[]  <- 숫자저장[100]
#변수[] 주소 -> 숫자저장[100]
```

```python
num
#num이 가리키는 값 출력
```

```python
print(num)
```

```python
# 함수 : 기능, 도구
id(num) #변수 주소값 확인 
# id(num) 출력 : 파이썬에서 부여한 매모리 주소값
```

```python
num1 = 1000
num2 = 2000
```

```python
print( id(num1) ) #print 안하면 마지막 함수값만 보임
print( id(num2) ) #object(num2) -> 객체
# 파이썬의 특징 숫자를 특정 메모리 묶음을 만들어놓고 하나로 본다 치면, 묶음 안에 변수, 함수들 가지고 있음.
# 숫자하나를 저장하는데 첫번째칸 2000, 두번칸 함수1, 세번째칸 함수2 이런식으로 처리됨. 이 묶음 형태를 가리켜서 객체라고함
# [2000|객체1|객체2]
# 파이썬에서는 모든 데이터를 객체형식으로 저장하게됨. 
# 숫자도 객체, 문자열도 객체, 논리값도 객체
```

```python
# 파이썬의 데이터(숫자, 문자열, 논리값)는 객체형식으로 저장
# 변수가 가리키는 메모리는 객체의 메모리 주소
# 객체 -> 변수, 함수들이 있음
# 변수를 알면 데이터 참조와 함수를 사용할 수 있다
# 변수.함수이름
num3 = 10   # 정수            #정수 저장방법 (객체)    ]
num4 = 3.14 # 실수            # 실수 저장방법         ]   => 자료형 Type
str_ten = 'ten' # 문자열 저장  # 물자열 저장 방법      ]
#저장 방법이 다르다는 것은 메모리 구성이 다르다는 이야기, 연산이 달라진다.
```

```python
# 자료형 확인 : type() 함수 사용
print(type(num3)) #print를 사용하면 <타입(class), '타입(Int)'>
print(type(num4))
print(type(str_ten))
```

```python
var = 100    # int type
var = 3.14   # float type
var = 'string' # str type
# 최종 string type이고, 주소값만 바꾸는 것이기 때문에 오류 안남
```

```python
# 논리값 : True / False
# True 
# false # <- 소문자는 식별자로 인식, 대소문자 구분

check_1 = True
check_2 = False
```

```python
print(num3, '+', num4, '=', num3 + num4) # 10 + 3.14 = 13.14 # 숫자+문자열은 X
print(num1, '+', num2, '=', num1 + num2) # 1000 + 2000 = 3000 
# 변수간의 연산, 사칙연산 가능
```

```python
# 문자열 생성
# 'string' 
# "string"
# '''string'''
# """string"""
str_1 = 'hello~' # 데이터 출력 후 개행(new line)
str_2 = "안녕하세요" # 데이터 출력 후 개행(new line)

# ''' ''' => 개행(new line,줄바꿈)도 문자열로 인식
# """ """ => 개행(new line,줄바꿈)도 문자열로 인식
str_3 = '''첫번째 라인
두번째 라인
세번째 라인
''' #개행
#개행
str_4 = """
selec *
from emp
where empno=2111
"""
```

```python
print(str_1)
print(str_2)
print(str_3)
print(str_4)
```

```python
# 문자열 연산 : +, *

# + : 문자열 + 문자열 => 앞 뒤 문자열을 연결해서 새로운 문자열 생성
print('hello~' + ' Python') # 문자열 이외 다른 형식은 안됨.

# * : 몇번 반복 붙이기
print('hello!' * 5)
```
