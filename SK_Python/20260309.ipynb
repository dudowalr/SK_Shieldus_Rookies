클래스의 상속

```python
# 사칙연산을 하는 클래스 : 클래스이름, 변수, 함수, 초기화 함수
# 클래스 이름 : 대문자로 시작 => FourCal
# 변수 : 숫자 두개
# 함수(메서드) : +, -, *, /

class FourCal:
    # 클래스의 멤버 : 변수, 함수

    # 변수 초기화 => 초기화 함수
    def __init__(self, first, second):
        self.first=first
        self.second=second

    # 함수
    # 클래스에서 함수를 사용할 때는 반드시 self 매개변수를 사용
    # self -> 실행 시점점의 함수를 포함하는 객체 자신을 가르키는 변수
    def setData(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second
    
    def sub(self):
        return self.first - self.second
    
    def mul(self):
        return self.first * self.second
    
    def div(self):
        return self.first / self.second
```

```python
# 클래스의 상속 : 새로운 클래스를 생성
# 기존의 클래스의 맴버(변수, 메서드)들을 받아서 새로운 클래스를 생성
# 오버라이딩 : 메서드를 수정(처리블럭)해서 사용
# MoreFourCal : 새로운 클래스 정의 => pow  제곱 구해주는 함수, div 메서드를 수정
class MoreFourCal(FourCal):
    #pass
    def pow(self):
        return self.first**self.second
    
    # 오버라이딩 : 함수의 이름과 매개변수는 그대로 하고 처리부분을 수정해서 정의
    def div(self):
        if self.second == 0 :
            return 0
        return self.first/self.second
```

```python
mfc1 = MoreFourCal(10, 0)
```

```python
#mfc1.pow()
mfc1.div()
```
