```python
import pandas as pd
```

```python
# 결측치 처리 
data={
    'Name': ['Alice', 'Bob', None],
    'Age': [25, None, 30],
    'Score': [85, 90, None]
    }
df = pd.DataFrame(data)
```

```python
df.info()
# 결측치 확인
df.isnull()
```

```python
# 결측치가 있는 행을 삭제
df_missing_rows = df.dropna()
print(df_missing_rows)
```

```python
# 결측치가 있는 열을 삭제
df_missing_cols = df.dropna(axis=1)
print(df_missing_cols)
```

```python
# 결측치에 특정값으로 대체

# 나이값을 평균값으로 대체
df['Age'] = df['Age'].fillna(df['Age'].mean())

# 점수값을 중간값으로 대체
df['Score'] = df['Score'].fillna(df['Score'].median())

# 이름값을 최빈값으로 대체
df['Name'] = df['Name'].fillna(df['Name'].mode()[0])

df
```

```python
#이상치 처리
data = {
    'score' : [70, 75, 80, 85, 90, 150]
}

df = pd.DataFrame(data)
df
```

```python
#IQP
Q1 = df['score'].quantile(0.25)
Q3 = df['score'].quantile(0.75)
IQR = Q3-Q1
print(Q1, Q3, IQR)
```

```python
low_ = Q1 - (1.5 * IQR)
up_ = Q3 + (1.5 * IQR)
print(f'이상치의 범위 {low_}보다 작은값, {up_}보다 큰값')
```

```python
# 이상치 값 찾기 
df_1 = df[ (df['score']<low_) | (df['score']>up_) ]
df_1
```

```python
# 이상치를 제거한 데이터
df_2 = df[(df['score']>=low_) & (df['score']<=up_)]
df_2
```

```python
# 이상치 데이터를 변경
df['score'] = df['score'].astype(float)
df.loc[df['score'] < low_, 'score'] = low_
df.loc[df['score']>up_, 'score'] = up_
df
```

```python
# 중복 데이터 처리 : 확인 -> 
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'Age': [25, 30, 25, 35, 30],
    'Score': [85, 90, 85, 95, 90]
}

df = pd.DataFrame(data)
df
```

```python
# 중복된 행을 삭제
df_clearned = df.drop_duplicates()
df_clearned
```

```python
df_clearned2 = df.drop_duplicates(keep='last')
df_clearned2
```

```python
df_clearned3 = df.drop_duplicates(subset=['Name'])
df_clearned3
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
```

```python
data1 = pd.DataFrame(
    {
        'original' : [50, 80, 90, 100, 160]
    }
    )
data2 = pd.DataFrame(
    {
        'Fruits' : ['사과', '배', '포도', '사과', '포도']
    }
)
```

```python
# 표준화
scaler_1 = StandardScaler()
data1['scale_1'] = scaler_1.fit_transform(data1[['original']])

# 정규화
scaler_2 = MinMaxScaler()
data1['scaled_2'] = scaler_2.fit_transform(data1[['original']])
data1
```
