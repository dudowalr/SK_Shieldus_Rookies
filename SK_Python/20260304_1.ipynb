### 연습문제 1
- 과목별 점수, 총점, 평균점수 출력
- 과목 : 국어, 영어, 수학

```python
k_score = int( input('국어 점수를 입력하세요') )   # 100 -> '100'
print('kor_score :', k_score, type(k_score))
print(int(k_score)-10)   # str + int  => int(str) -> int => int(str) + int
```

```python
# 변수
#kor_score = 100
#eng_score = 70
#math_score = 80
```

```python
# 변수
kor_score = int( input('국어 점수를 입력하세요') )
eng_score = int( input('영어 점수를 입력하세요') )
math_score = int( input('수학 점수를 입력하세요') )
```

```python
# 연산 : 총점, 평균
sum_score = kor_score + eng_score + math_score
avg_score = sum_score / 3
```

```python
# 출력
print('성적표')
print('-'*30)
print('국어 :', kor_score, '점')
print('영어 : ' + str(eng_score) + ' 점' )
print(f'수학 : {math_score} 점')
print('-'*30)
print(f'총점 : {sum_score} 점')
print(f'평균 : {avg_score} 점')
```
