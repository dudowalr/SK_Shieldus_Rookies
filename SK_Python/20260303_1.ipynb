```python
# 문자열 생성
# 'string'
# "string"
# '''string'''
# """string"""
str_1 = 'hello~'
str_2 = "안녕하세요"
# ''' ''' => 개행(줄바꿈)도 문자열로 인식
# """ """ => 개행(줄바꿈)도 문자열로 인식
str_3 = '''첫번째 라인
두번째 라인
세번째 라인
'''
str_4 = """
select *
from emp
where empno=2111
"""
```

```python
print(str_1) # 데이터 출력 후 개행
print(str_2)
print(str_3)
print(str_4)
```

```python
print(type(str_3), type(str_4))
```

```python
# 문자열 연산 : +, *
# + : 문자열 + 문자열 => 앞 뒤 문자열을 연결해서 새로운 문자열 생성
print('hello~' + ' Python')
# * : 몇번 반복 붙이기
print('hello!'*5)
print('-'*30)
```
