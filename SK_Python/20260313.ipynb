```python
import pandas as pd
```

```python
#고객 정보 데이터 프레임
customers = pd.DataFrame({
    'customer_id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

# 주문 정보 데이터프레임
orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 4],
    'product': ['Laptop', 'Tablet', 'Smartphone']
})
```

```python
customers
```

```python
# merge :열단위 데이터 병학 => 옆으로 열을 붙여서 새로운 표를 만든다
#공통된 customer_id로 데이터 병합
result1 = pd.merge(customers, orders, on='customer_id', how='inner')
result1
```

```python
result2 = pd.merge(customers, orders, on='customer_id', how='right')
result2
```

```python
# 1월 매출 데이터
sales_jan = pd.DataFrame({
'month': ['January', 'January', 'January'],
'product': ['Laptop', 'Tablet', 'Smartphone'],
'sales': [100, 150, 200]
})
# 2월 매출 데이터
sales_feb = pd.DataFrame({
'month': ['February', 'February', 'February'],
'product': ['Laptop', 'Tablet', 'Smartphone'],
'sales': [120, 160, 220]
})
```

```python
# Row-wise Concatenation (1월과 2월 데이터를 이어붙임)
sales_combined = pd.concat([sales_jan, sales_feb], axis=0)
# 결과 출력
print(sales_combined)
```

```python
# 새로운 df
data_1 = df[['번호', '영업상태명', '사업장명', '소재지전체주소', '도로명전체주소']]
data_1
```

```python
# 소재지 기준으로 값 NaN인 값을 삭제
data_1 = data_1.dropna(subset=['소재지전체주소'])
data_1.info()
```

```python
# 소재지 기준으로 '서울특별시'를 포함하는 행만 남기는 처리
data_2 = data_1[data_1['소재지전체주소'].str.contains('서울특별시')]
data_2.info()
```

```python
data_2.head()
```

```python
# 사업장 명 => '스타벅스' 
data_3 = data_2[data_2['사업장명'].str.contains('스타벅스')]
data_3.info()
```

```python
data_4 = data_3[data_3['영업상태명'] != '폐업']
data_4.info()
```

```python
data_4
```

```python
# 열을 추가 : 시군구
# 구 별로 스타벅스 매장이 몇개가 있는지 
data_4.loc[:, '시군구'] = data_4['소재지전체주소'].str.split().str[1]
data_4
```

```python
data_5 = data_4.groupby('시군구').size().reset_index(name='사업장수')
data_5
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Batang'
plt.bar(
    data_5['시군구'],
    data_5['사업장수']
)
plt.xticks(rotation=45)
plt.title('서울특별시 스타벅스 사업장 분포')
plt.xlabel('시군구')
plt.ylabel('사업장수')
```
