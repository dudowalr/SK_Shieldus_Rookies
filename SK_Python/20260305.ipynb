```python
# if 문 : 분기 => ~라면, ~ 아니라면 , 조건에 따라 처리방식을 정의하는 방법
# 조건식을 잘 만든것이 중요!
# 성인 판단 => 성인이라면 입장, 아니라면 킥!!

age = 13    # 사용자 입력 데이터, DB, file, ....

# 성인=>환영메시지, 미성년=>킥
if age >= 18 : # 조건의 결과 => 논리값
    print('환영합니다.')
    print('서버에 접속합니다. 게임을 시작합니다.')
else : # False 일때 실행
    print('킥~!!!. 오면 안돼요!!!')


```

```python
# 짝수, 홀수 판별
# 조건식 : num%2 => 0(짝수), 1(홀수)
# num%2 == 0 => 짝수, num%2 == 1 => 홀수
num = int(input('숫자를 입력하세요'))

if num%2 == 1 :
    print('홀수')
else :
    print('짝수')
```

```python
# if 문의 충첩
# 양수, 음수, 0 구별
# 양수 : num > 0
# 음수 : num < 0
# 0 : num>0 아니고 num<0

num = 0

if num>0 :
    print('양수')
else : 
    # 음수, 0 
    if num<0 :
        print('음수')
    else : # 0보다 크지도 않고 작지도 않은 상태
        print('0')
```

```python
# 반복문 : while
# while condition(True/False) :
#     반복 구문 ...

# 1~5 까지의 숫자를 출력
# 반복 작업 => print()

cnt = 1

while cnt < 6 :
    print(cnt)
    cnt = cnt + 1
```

```python
# 1~10 까지 출력, 홀수만 출력
cnt = 1 
while cnt < 11 :
    if cnt%2==1:
        print(cnt)
    cnt += 1 # cnt = cnt + 1
```

```python
# continue
# 반복문 내에서 이 키워드를 만나면 아래 코드들은 실행하지 않고 블록을 종료
#  => 조건식으로 이동
cnt=0
while cnt<10:
    cnt += 1
    if cnt%2==0:
        continue
    print(cnt)
```

```python
# break : 반복문 내에서 이 키워드를 만나면 반복문이 종료
cnt = 0
while True :
    cnt += 1 
    print(cnt)
    if cnt > 4:
        break
```

```python
# 5명의 학생의 영어 시험 점수 저장 => list
# 반복하면서 60 점이 이상 PASS, 아니면 FAIL
# 총점, 평균
scores = [80, 50, 70, 100, 55]
num = 0     # 학생 일련번호
sum = 0
for s in scores:
    num += 1
    # pass check
    msg = 'PASS' if s>=60 else 'FAIL'
    print(num, s, msg)
    sum = sum + s   # 합을 더해나가는 연산 식

print('-'*30)
print('총점 :', sum)
print('평균 :', sum/len(scores))
```

```python
# range 함수를 이용해서 구구단을 만들어보자
# 2단을 만들어보자
# 2X1=2, 2X2=4, ..., 2X9=18
for num in range(1, 10):
    print(f'2X{num}={2*num}')    # 2 X {} = {2*}

for num in range(1, 10):
    print(f'3X{num}={3*num}')    # 2 X {} = {2*}

for num in range(1, 10):
    print(f'4X{num}={4*num}')    # 2 X {} = {2*}
```

```python
# 단 표현
for dan in range(2,10):
    print(f'{dan} 단')
    for i in range(1, 10):
        print(f'{dan}X{i}={dan*i}', end='  ')
    print()
```

```python
# 리스트의 컴프리핸션 => 새로운 리스트 생성
# [표현식 for var in list if 조건식]
list_c = [1,2,3,4]  # *3 => [3,6,9,12]
list_result = []

for num in list_c:
    if num%2==1:
        print(num)
        list_result.append(num*3)

print(list_result) 
```

```python
list_result2 = [num*3 for num in list_c if num%2==1]
print(list_result2)
```

함수

```python
# 함수 : 데이터를 받고 처리해서 결과를 반환하는 코드들의 집합
# 처리가 필요할 때 함수를 호출해서 사용
# 공통코드들을 줄이고 가독성 향상, 유지보수, 확장성이 좋아짐
# def 함수이름(매개변수1, 매개변수2, ...) :
#     처리해야할 코드들...
#     return 반환data(파이썬에서 제공하는 모든 데이터 가능)
```

```python
# 함수 정의 : 두개의 숫자를 받아서 더하기 연산을 한 후 결과를 반환하는 함수
# def
# 이름 : add
# 매개변수(선택적) : (n1, n2)
# 처리 : 덧샘 => n1 + n2
# 반환(선택적) : return n1+n2
# 매개변수, 반환
def add(n1,n2) :
    result = n1 + n2
    return result
```

```python
# 함수의 호출(사용) => 이름과 괄호안에 배개변수를 넣어서 호출
# 매개변수가 필요없는 경후 => ()
add(10, 30)
```

```python
# 함수의 타입 => 일급 객체다 => 변수가 함수를 카르킨다!
print(type(add))
```

```python
# 매개변수는 있고, 반환이 없는 함수
# 숫자 두개를 받아서 덧샘한 결과를 출력하는 함수
# def 
# 이름 add2
# 매개변수 (n1, n2)
# 처리 print(n1+n2)
# return 은 없다!
def add2(n1,n2):
    print('n1+n2 =', n1+n2)
```

```python
add2(10,30)
```

```python
result_3 = add2(100, 200)
print(result_3)
```
